```
# @title importar a pandas e carregar os arquivos CSV

import pandas as pd

# Carregar os arquivos CSV
vendedores_df = pd.read_csv('/content/Vendedores.csv')
fornecedores_df = pd.read_csv('/content/Fornecedores.csv')
transportadoras_df = pd.read_csv('/content/Transportadoras.csv')
vendas_globais_df = pd.read_csv('/content/Vendas Globais.csv')

# Exibir as primeiras linhas de cada dataframe para análise
vendedores_df.head(), fornecedores_df.head(), transportadoras_df.head(), vendas_globais_df.head()
```

```
# @title Pergunta 1: Quem são os 10 maiores clientes em termos de vendas ($)?

# Agrupar as vendas por cliente e somar o valor total das vendas, ordenando em ordem decrescente
top_10_clientes = vendas_globais_df.groupby("ClienteNome")["Vendas"].sum().nlargest(10).reset_index()
top_10_clientes
```

```
# @title Pergunta 2: Quais os três maiores países em termos de vendas ($)?

# Agrupar as vendas por país e somar o valor total das vendas, ordenando em ordem decrescente
top_3_paises = vendas_globais_df.groupby("ClientePaís")["Vendas"].sum().nlargest(3).reset_index()
top_3_paises
```

```
# @title Pergunta 3: Quais as categorias de produtos que geram maior faturamento (vendas $) no Brasil?

# Filtrar vendas apenas para o Brasil e agrupar por categoria, somando o valor das vendas
faturamento_categorias_brasil = (
    vendas_globais_df[vendas_globais_df["ClientePaís"] == "Brazil"]
    .groupby("CategoriaNome")["Vendas"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
faturamento_categorias_brasil
```

```
# @title Pergunta 4: Qual a despesa com frete envolvendo cada transportadora?

# Agrupar as vendas por transportadora e somar os custos de frete
despesa_frete_transportadora = (
    vendas_globais_df.groupby("TransportadoraID")["Frete"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

# Mesclar com o dataframe de transportadoras para obter os nomes
despesa_frete_transportadora = despesa_frete_transportadora.merge(transportadoras_df, on="TransportadoraID")
despesa_frete_transportadora = despesa_frete_transportadora.rename(columns={"Frete": "Despesa com Frete"})
despesa_frete_transportadora
```

```
# @title Pergunta 5: Quais são os principais clientes do segmento "Calçados Masculinos" (Men’s Footwear) na Alemanha?

# Filtrar para o segmento "Calçados Masculinos" na Alemanha e agrupar por cliente, somando as vendas
top_clientes_mens_footwear_ger = (
    vendas_globais_df[
        (vendas_globais_df["CategoriaNome"] == "Men´s Footwear") &
        (vendas_globais_df["ClientePaís"] == "Germany")
    ]
    .groupby("ClienteNome")["Vendas"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
top_clientes_mens_footwear_ger
```

```
# @title Pergunta 6: Quais os vendedores que mais dão descontos nos Estados Unidos?

# Filtrar as vendas para os Estados Unidos e agrupar por vendedor, somando os descontos
vendedores_descontos = (
    vendas_globais_df[vendas_globais_df["ClientePaís"] == "USA"]
    .groupby("VendedorID")["Desconto"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

# Mesclar com o dataframe de vendedores para obter os nomes
vendedores_descontos = vendedores_descontos.merge(vendedores_df, on="VendedorID")
vendedores_descontos
```

```
# @title Pergunta 7: Quais os fornecedores que dão a maior margem de lucro no segmento "Vestuário Feminino"?

# Filtrar para o segmento "Vestuário Feminino" e criar uma cópia
womens_wear = vendas_globais_df[vendas_globais_df["CategoriaNome"] == "Womens wear"].copy()

# Calcular a margem de lucro como diferença entre vendas e custo
womens_wear["Margem_Lucro"] = womens_wear["Vendas"] - womens_wear["Vendas Custo"]

# Agrupar por fornecedor e somar a margem de lucro
fornecedores_margem = (
    womens_wear.groupby("FornecedorID")["Margem_Lucro"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

# Mesclar com o dataframe de fornecedores para obter os nomes
fornecedores_margem = fornecedores_margem.merge(fornecedores_df, on="FornecedorID")
fornecedores_margem
```

```
# @title Pergunta 8: Quanto foi vendido no ano de 2009, e como foi o faturamento de 2009 a 2012?

# Converter a coluna "Data" para o formato de data
vendas_globais_df["Data"] = pd.to_datetime(vendas_globais_df["Data"], format="%d/%m/%Y", errors='coerce')

# Filtrar as vendas por ano e agrupar para obter o faturamento anual
vendas_anuais = (
    vendas_globais_df[vendas_globais_df["Data"].dt.year.between(2009, 2012)]
    .groupby(vendas_globais_df["Data"].dt.year)["Vendas"]
    .sum()
    .reset_index()
    .rename(columns={"Data": "Ano", "Vendas": "Faturamento"})
    .sort_values(by="Ano")
)
vendas_anuais
```

```
# @title Pergunta 9: Quais são os principais clientes (vendas $) do segmento “Calçados Masculinos” (Men ́s Footwear) na Alemanha?

# Filtrar para o segmento "Calçados Masculinos" na Alemanha e agrupar por cliente, somando as vendas
top_clientes_mens_footwear_ger = (
    vendas_globais_df[
        (vendas_globais_df["CategoriaNome"] == "Men´s Footwear") &
        (vendas_globais_df["ClientePaís"] == "Germany")
    ]
    .groupby("ClienteNome")["Vendas"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
top_clientes_mens_footwear_ger
```

```
# @title Pergunta 10: Quais os países nos quais mais se tiram pedidos?

# Agrupar por país e contar o número total de pedidos
total_pedidos_pais = (
    vendas_globais_df.groupby("ClientePaís")["PedidoID"]
    .count()
    .sort_values(ascending=False)
    .reset_index()
    .rename(columns={"PedidoID": "Total de Pedidos"})
)
total_pedidos_pais
```
